```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### First Plot using matplotlib

```python
df = pd.read_csv('iris.csv')
```

```python
df.sample(10)
```

```python
df.shape
```

```python
#Line Plot
plt.plot(df['sepal.length'])
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()
```

## Multiple plot in a Figure and Legends

```python
plt.figure(figsize=(10,6))
plt.plot(df['sepal.length'], label ='Sepal Lenght' )
plt.plot(df['sepal.width'], label ='Sepal Widht' )
plt.plot(df['petal.length'], label ='Petal Lenght' )
plt.plot(df['petal.width'], label ='Petal Width' )
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Length of Sepal and Petal')
plt.legend()
plt.show()
```

## Changing the Color of Line plot

```python
plt.figure(figsize=(15,10))
plt.plot(df['sepal.length'],'r*', label ='Sepal Lenght' )
plt.plot(df['sepal.width'], 'go', label ='Sepal Widht' )
plt.plot(df['petal.length'], 'b^', label ='Petal Lenght' )
plt.plot(df['petal.width'], 'ys', label ='Petal Width' )
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Length of Sepal and Petal')
plt.legend()
plt.show()
```

## Changing the size of line plot

```python
plt.figure(figsize=(10,5))
plt.plot(df['sepal.length'],'r', linewidth=5, label ='Sepal Lenght' )
plt.plot(df['petal.width'], 'y', label ='Petal Width', )
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Length of Sepal and Petal')
plt.legend()
plt.show()
```

## Buliding a Subplot

```python
plt.figure(figsize=(15,8))
plt.subplot(221)
plt.grid()
plt.xlabel('index')
plt.ylabel('Sepal Lenght')
plt.plot(df['sepal.length'],'r', label='Sepal Lenght')
plt.legend()

plt.subplot(222)
plt.grid()
plt.xlabel('index')
plt.ylabel('Sepal Width')
plt.plot(df['sepal.width'], 'g', label = 'Sepal Widht')
plt.legend()

plt.subplot(223)
plt.grid()
plt.xlabel('index')
plt.ylabel('Petal Lenght')
plt.plot(df['petal.length'],'b', label='Petal Lenght')
plt.legend()

plt.subplot(224)
plt.grid()
plt.xlabel('index')
plt.ylabel('Petal Width')
plt.plot(df['petal.width'],'y', label='Petal Width')
plt.legend()
plt.show()
```

## Ploting Histogram

```python
# X axis is min lenght to max lenght and y axis is total count. So below graph represnts 5.3  to 5.8 data is the top most count.
# bins represnts total division to min to max lenght
plt.hist(df['sepal.length'],  color='r')
plt.show()
```
