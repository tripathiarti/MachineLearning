### First Plot using matplotlib

```python
arr = np.array([2,4,8,16])
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr)
plt.show()
```

## Using Label and Title

```python
arr1 = np.arange(0,100,0.1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr1,arr1**2)
plt.show()
```

## Multiple plots in a figure

```python
arr2 = np.arange(0,10,1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr2,arr2*2)
plt.plot(arr2,arr2*3)
plt.plot(arr2,arr2*4)
plt.show()
```

## Changing color of plot

```python
arr2 = np.arange(0,10,1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr2,arr2*2,'r')
plt.plot(arr2,arr2*3, 'g')
plt.show()
```

## Changing Shapes and using legend

```python
arr2 = np.arange(0,10,1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr2,arr2*2,'r^',label='mul-2')
plt.plot(arr2,arr2*3, 'go', label='mul-3')
plt.plot(arr2,arr2*4, 'bs', label='mul-4')
plt.legend()
plt.show()
```

## Changing the line width

```python
arr2 = np.arange(0,10,1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr2,arr2*2,label='mul-2')
plt.plot(arr2,arr2*3,linewidth=5, label='mul-3')
plt.legend()
plt.show()
```

## Changing the figuresize

```python
plt.figure(figsize=(10,6)) #figuresize
arr2 = np.arange(0,10,1)
plt.xlabel("X-label-sample")
plt.ylabel("Y-label-sample")
plt.title("Lineplot")
plt.plot(arr2,arr2*2,label='mul-2')
plt.plot(arr2,arr2*3,linewidth=5, label='mul-3')
plt.legend()
plt.show()
```

## Building subplots

```python
plt.subplot(211) # 2- rows, 1- column, 1 - figure number
plt.grid()
plt.plot(arr2, arr2**2)
plt.plot(arr2, arr2*2)

plt.subplot(212)
plt.plot(arr2, arr2**3)

plt.show()
```

```python
plt.subplot(121) # 1- rows, 2- column, 1 - figure number
plt.grid()
plt.plot(arr2, arr2**2)
plt.plot(arr2, arr2*2)

plt.subplot(122)
plt.plot(arr2, arr2**3)

plt.show()
```

```python
import pandas as pd
```

```python
titanicdf= pd.read_csv('titanic.csv')
```

```python
titanicdf.shape
```

```python
titanicdf.sample(10)
```

```python
len(titanicdf['Fare'])
```

```python
plt.plot(titanicdf['Fare'])
```

## Plotting Histogram

```python
#Univariate analysis
plt.hist(titanicdf['Age'], bins=10, color='g')
plt.show()
```

```python
plt.grid()
plt.hist(titanicdf['Fare'], bins=10)
plt.show()
```

## Plotting Scatterplot

```python
plt.figure(figsize=(10,6))
plt.scatter(x= 'Age', y = 'Fare', data = titanicdf, color='b')
plt.show()
```

```python
titanicdf.corr()
```

```python
rows = []
for loopvar, subset in titanicdf.groupby('Sex'): #group by sex
    total_survived = subset['Survived'].sum()
    rows.append([loopvar,len(subset),total_survived])
newdf = pd.DataFrame(rows,columns=["Tag","OverallCount","SurvivedCount"])
```

```python
newdf
```

## Bar graph

```python
plt.figure(figsize=(10,6))
arr = np.arange(2)
plt.bar(arr,newdf['SurvivedCount'], width=0.1, align='center',label= "SurvivedCount", tick_label=newdf['Tag'].unique())
plt.bar(arr+0.1, newdf['OverallCount'], width=0.1, align='center',label= "OverallCount",tick_label=["Female","male"])
plt.title("Male Vs Female titanic Survival")
plt.legend()
plt.show()
```

```python
titanicdf['Sex'].value_counts()
```

```python
plt.pie(titanicdf['Sex'].value_counts(), labels=titanicdf['Sex'].unique())
plt.title("Test")
plt.show()
```

# Seaborn Demo

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(5,6))
sns.barplot(x="day", y="total_bill",data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="day", y="total_bill",hue='time', data=df)
plt.show()
```

```python
sns.distplot(df['total_bill'],bins=10, kde=False) #kernal Density estimation
plt.show()
```

```python
sns.distplot(df['total_bill'])
plt.show()
```

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.distplot(df[''total_bill''])
sns.distplot(df['tip'])
plt.show()
```

```python
sns.scatterplot(df['total_bill'], df['tip'])
```

```python
sns.scatterplot(df['total_bill'], df['tip'], hue=df['smoker'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.grid()
sns.scatterplot(df['total_bill'], df['tip'], hue=df['smoker'], style=df['day'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.grid()
sns.scatterplot(df['total_bill'], df['tip'], hue=df['day'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot('total_bill', 'tip', hue='day', data = df)
plt.show()
```

```python
sns.lmplot(x= 'total_bill', y = 'tip', hue='smoker', data = df)
```

```python
sns.regplot(df['total_bill'], df['tip'])
plt.show()
```
